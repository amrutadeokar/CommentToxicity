```python
from tensorflow.keras.layers import TextVectorization
```

```python
x = df['comment_text']
y = df[df.columns[2:]].values
```

```python
MAX_FEATURES =  200000        # numbers of words in the vocab
```

```python
MAX_SEQUENCE_LENGTH = 1800
```

```python
vectorizer = TextVectorization(max_tokens = MAX_FEATURES,
                               output_sequence_length = MAX_SEQUENCE_LENGTH ,
                               output_mode = 'int')
```

```python
vectorizer.adapt(x.values)
```

```python
vectorizer('Hello World, Life is great')
```

```python
vectorizer.get_vocabulary()
```

```python
vectorized_text = vectorizer(x.values)
```

```python
print(vectorized_text.shape)
```

```python
# MCSHBAP - map, chache, shuffle, batch, prefetch from_tensor_slices, list_file
dataset = tf.data.Dataset.from_tensor_slices((vectorized_text, y))
dataset = dataset.cache()
dataset = dataset.shuffle(160000)
dataset = dataset.batch(16)
dataset = dataset.prefetch(8)    # helps bottlenecks
```

```python
batch_x, batch_y = dataset.as_numpy_iterator().next()
```

```python
train = dataset.take(int(len(dataset)*.7))
val = dataset.skip(int(len(dataset)*.7)).take(int(len(dataset)*.2))
test = dataset.skip(int(len(dataset)*.9)).take(int(len(dataset)*.1))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import  Embedding, LSTM, Dropout, Bidirectional, Dense
```

```python
model = Sequential()
# Create the embedding layer
model.add(Embedding(input_dim = MAX_FEATURES + 1, output_dim = 32))
# Bidirectional LSTM Layer
model.add(Bidirectional(LSTM(32, activation = 'tanh')))
# Feature extractor Fully connected layer
model.add(Dense(128, activation = 'relu'))
model.add(Dense(256, activation = 'relu'))
model.add(Dense(128, activation = 'relu'))
# Final layer
model.add(Dense(6, activation = 'sigmoid'))
```

```python
model.compile(loss = 'binary_crossentropy', optimizer = 'adam')
```
